```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./Results/ICCV/processed.csv")
data
```

```python
data_cifar_gmm = data[(data["Model"] == "densenet") & (data["ID dataset"] == "cifar100") & (data["Monitor"] == "ReAct_Energy")]
data_cifar_gmm
```

```python
N = data_cifar_gmm.shape[0]
n = int(N / 9)
```

```python
AUPR = data_cifar_gmm["AUPR"].to_numpy()
ranking = []
for i in range(data_cifar_gmm.shape[0]):
    array = data_cifar_gmm["AUPR"][data_cifar_gmm["OOD dataset"] == data_cifar_gmm["OOD dataset"].iloc[i]]
    ranks = np.array(-array).argsort().argsort()
    ranking.append(ranks[i%n] + 1)
ranking = np.array(ranking)
```

```python
ranking_other = []
for i in range(data_cifar_gmm.shape[0]):
    j = i%n
    ranking_other.append((np.sum(ranking[np.arange(j, N, n)]) - ranking[i])/8)
ranking_other = np.array(ranking_other)
```

```python
for i in range(9):
    print(np.argmin(ranking_other[n*i:n*(i+1)]))
```
